```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import rotate, shift, zoom
```

```python
img = np.load(r"D:\extracteddata\018_010\CT.npy")
mask_l = np.load(r"D:\extracteddata\018_010\parotid_l_mask.npy")
mask_r = np.load(r"D:\extracteddata\018_010\parotid_r_mask.npy")
```

```python
rot = rotate(img, angle=15, axes=(1,2),reshape=False, mode='nearest')
rmask_l = rotate(mask_l, angle=15, axes=(1,2),reshape=False, mode='constant',cval=0)
rmask_r = rotate(mask_r, angle=15, axes=(1,2), reshape=False, mode='constant', cval=0)
```

```python
merged_mask = mask_l + mask_r
rmerged_mask = rmask_l + rmask_r
```

```python
%matplotlib widget

select = 129

fig, ax = plt.subplots(1,2,figsize=(12,6))
ax[0].imshow(img[select,:,:],cmap='gray')
ax[0].imshow(merged_mask[select,:,:],alpha=0.5)
ax[1].imshow(rot[select,:,:],cmap='gray')
ax[1].imshow(rmerged_mask[select,:,:],alpha=0.5)
```

```python
shifted = shift(img, shift=(0,40,15), mode='nearest')
smask_l = shift(mask_l, shift=(0,40,15), mode='constant', cval=0)
smask_r = shift(mask_r, shift=(0,40,15), mode='constant', cval=0)

smerged_mask = smask_l + smask_r
```

```python
%matplotlib widget

select = 129

fig, ax = plt.subplots(1,2,figsize=(12,6))
ax[0].imshow(img[select,:,:],cmap='gray')
ax[0].imshow(merged_mask[select,:,:],alpha=0.5)
ax[1].imshow(shifted[select,:,:],cmap='gray')
ax[1].imshow(smerged_mask[select,:,:],alpha=0.5)
```

```python
zoomed = zoom(img, zoom=0.8, mode='nearest')
zmask_l = zoom(mask_l, zoom=0.8, mode='constant', cval=0)
zmask_r = zoom(mask_r, zoom=0.8, mode='constant', cval=0)

zmerged_mask = zmask_l + zmask_r
```

```python
%matplotlib widget

select = 129
adjust = 1

fig, ax = plt.subplots(1,2,figsize=(12,6))
ax[0].imshow(img[select,:,:],cmap='gray')
ax[0].imshow(merged_mask[select,:,:],alpha=0.5)
ax[1].imshow(zoomed[100,:,:],cmap='gray')
ax[1].imshow(zmerged_mask[100,:,:],alpha=0.5)
```
